# converted o and 1 values of senior citizen to Yes/No

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(20)
```

```python
ax = sns.countplot(x = "Churn" , data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels= gb.index,autopct = "%1.2f%%")
plt.title("Percentage of Churn Customer")
plt.show()
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender",data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen",data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by SeniorCitizen")
plt.show()
```

```python
grouped = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize=True).unstack() * 100

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(5, 5))
grouped.plot(kind="bar", stacked=True, colormap="coolwarm", ax=ax)

# Add labels
for i, (index, row) in enumerate(grouped.iterrows()):
    bottom = 0
    for churn_status in grouped.columns:
        percentage = row[churn_status]
        if percentage > 0:
            ax.text(i, bottom + percentage / 2, f"{percentage:.1f}%", ha="center", fontsize=12, color="white")
            bottom += percentage

# Customize chart
plt.title("Churn by Senior Citizen (Stacked bar chart)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")
plt.ylim(0, 100)  # Percentage scale

# Show plot
plt.show()
```

```python
cols = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
        'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3x3 grid
fig.suptitle("Countplots for Various Services", fontsize=16)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Generate countplots for each column
for i, col in enumerate(cols):
    sns.countplot(x=df[col], ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f"{col} Distribution")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=30)  # Rotate x-axis labels if needed

# Adjust layout for better readability
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# Customers with fiber optic internet, streaming services, and no security or backup services show higher churn rates. Tech support, online security, and backup services seem to reduce churn. Customers without internet service or extra phone lines are less likely to churn.
